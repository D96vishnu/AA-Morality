```python
import scrapy 
from scrapy.crawler import CrawlerProcess, CrawlerRunner
import os
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
class BBCSpider(scrapy.Spider):
    name = 'bbc_spider'
    
    def start_requests(self):
        bbc_url = "https://www.bbc.co.uk/search?q=alcoholics+anonymous&page="
        pages = 29
        urls = [bbc_url+str(page) for page in range(1,pages+1)]
        for url in urls:
            yield scrapy.Request(url = url, callback = self.parse_for_links)
    def parse_for_links(self, response):
        links = response.xpath('//a[contains(@class,"PromoLink")]/@href').extract()
        for url in links:
            yield response.follow(url, callback = self.parse_page)
    
    def parse_page(self, response):
        page_title = response.xpath('//h1[contains(@id,"main-heading")]/text()')
        page_title = page_title.extract_first().strip()
        
        page_data = response.xpath("//p/text()")
        page_data = [t.strip() for t in page_data.extract()]
        
        page_dict[page_title] = page_data
        
        
```

```python
page_dict = dict()
process = CrawlerProcess()
process.crawl(BBCSpider)
process.start()
```
